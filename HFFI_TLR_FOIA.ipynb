```python
import geopandas as gpd
import matplotlib.pyplot as plt
import shapely as shply
print(gpd.__version__)  # note this is a pretty old version
```

```python
gdf = gpd.read_file('2020-02-112 hffi data fy18.csv')
```

```python
gdf.head()
```

```python
type(gdf.project_address_latitude__c)
```

```python
#TODO - Need to convert latitude and longitude Series from strings to floats


#df["A"] = pd.to_numeric(df["A"], downcast="float")
gdf['float_lat'] = pd.to_numeric(gdf['project_address_latitude__c'], downcast='float')
gdf['float_long'] = pd.to_numeric(gdf['project_address_longitude__c'], downcast='float')


# creating a geometry column 
geometry = [shply.geometry.point.Point(xy) for xy in zip(gdf['float_lat'], gdf['float_long'])]


# Coordinate reference system : WGS84
crs = {'init': 'epsg:4326'}

# Creating a Geographic data frame 
gdf_2 = gpd.GeoDataFrame(gdf, crs=crs, geometry=geometry)
```

```python
# Download a cartographic shapefile of the 50 states, which will include coastline

import requests
url = 'https://www2.census.gov/geo/tiger/GENZ2016/shp/cb_2016_us_state_500k.zip'

with open('cb_2016_us_state_500k.zip', 'wb') as f:
    r = requests.get(url)
    f.write(r.content)
    
outlines = gpd.read_file('zip://cb_2016_us_state_500k.zip')
```

```python
# Plot the coastline data we just loaded
ax = geoplot.polyplot(outlines, 
                      projection = geoplot.crs.AlbersEqualArea(),
                      extent = (-123, 36.9, -121.4, 38.7))  # lower left, upper right


# Overlay the  points
geoplot.pointplot(gdf_2, 
                  projection = geoplot.crs.AlbersEqualArea(),
                  ax = ax,  # add it to the axes created above
                  s = 15)

ax.set_title("HFFI Loan Recipients")

#This map obviously needs a lot of work, but its fun to have something to look at!
plt.show()
```
