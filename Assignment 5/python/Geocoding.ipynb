# Geocoding

```python
import pandas as pd
import geopy
```

```python
df = pd.read_csv("hffi_data_FY20122018_with_geo_final.csv")
df.head()
len(df)
```

```python
df = df.loc[df['address']!=" "]
df.address.replace("\n", "")
df.head(100)
```

Continuously checking length after each operation to check how many lines are lost in the process

```python
len(df)
df.to_csv("hffi_data_FY20122018_beforegeocode.csv")
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="**armour@berkeley.edu**")
```

```python
# 1 - - create location column
df['location'] = df['address'].apply(geolocator.geocode, timeout=50)
df['point'] = df['location'].apply(lambda loc: tuple(loc.point) if loc else None)
df = df.loc[df['point'].notnull()]
```

```python
len(df)

## it is dropping most addresses with abbreviations, punctionation (N. Street) and PO Boxes
```

```python
df.head(100)
```

```python
# 2 - split point column into latitude, longitude and altitude columns
df['lat'] = df.apply(lambda x: x['point'][0], axis=1)
df['long'] = df.apply(lambda x: x['point'][1], axis=1)
df['alt'] = df.apply(lambda x: x['point'][2], axis=1)

df.head()
```

```python
len(df)
```

```python
df.to_csv("hffi_data_FY20122018_with_geo_final_2.csv")
```
