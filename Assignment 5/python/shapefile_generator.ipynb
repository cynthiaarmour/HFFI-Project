# Here, we add census-tract level health data to tract geometries

### Import our Data

```python
#We need to decide what state we want to map here

importFile = 'df_factory_outputs/kidneyDF.csv'
tractShapefile = 'shapefiles/tennessee/cb_2017_47_tract_500k.shp'
outFile = 'shapefiles/kidney_tennessee.shp'

# tractShapefiles maybe from here: 'http://www2.census.gov/geo/tiger/GENZ2015/shp/cb_2015_06_tract_500k.zip
```

```python
typesDict = {'TractFIPS': str}
df = pd.read_csv(importFile, dtype=typesDict)
```

```python
#get the census tract boundaries (for histrorical reasons, tract shapefile variable called coTracts)
coTracts = gpd.read_file(tractShapefile)
coTracts['TractFIPS'] = coTracts['GEOID']
print ('Nubmer of rows in our tracts dataframe: ', len(coTracts))

# merge our health dataframe with our census tracts dataframe on FIPS codes
gdf = coTracts.merge(df, on='TractFIPS', how='left')
print('Number of tracts for which we have health data: ', len(gdf))
print(type(gdf))

gdf.head()
```

```python
assert(len(gdf.loc[gdf['geometry'].isna()]) == 0)
```

## Write out Our File, now with Health Data + Tract Geographies


```python
gdf.to_file(outFile)
```
